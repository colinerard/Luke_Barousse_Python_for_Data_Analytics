# How are the top 5 skills for Data Analysts trending over time?

### Methodology

1. Aggregate skill counts monthly
2. Calculate relative importance of skills over all job postings for each month
3. Plot the monthly relative skill demand

### Librairies import and data loading

```python
# Load Librairies
import ast
import pandas as pd
import seaborn as sns
from datasets import load_dataset
import matplotlib.pyplot as plt

# Load Data
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset["train"].to_pandas()

# Clean data
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
df["job_skills"] = df["job_skills"].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x) 
```

### Calculations

```python
# focus on Data Analyst roles in the United States

df_DA_US = df[(df["job_title_short"] == "Data Analyst") & (df["job_country"] == "United States")].copy()
```

```python
# create a new column to highlight the month the job was posted

df_DA_US["job_posted_month_no"] = df["job_posted_date"].dt.month
```

```python
# break out the skills required for data analysis

df_DA_US_exploded = df_DA_US.explode("job_skills")
```

```python
# let's pivot the data we have to identify the number of skills demanded per month

df_DA_US_pivot = df_DA_US_exploded.pivot_table(index="job_posted_month_no", columns="job_skills", aggfunc="size", fill_value=0)

# adding a new row to calculate the total number of job postings mentioning each skill

df_DA_US_pivot.loc["Total"] = df_DA_US_pivot.sum()

# sort the dataframe for the top 5 most mentioned skills

df_DA_US_pivot = df_DA_US_pivot[df_DA_US_pivot.loc["Total"].sort_values(ascending=False).index]

# drop the total row

df_DA_US_pivot = df_DA_US_pivot.drop("Total")

df_DA_US_pivot
```

```python
# calculate how many data analyst jobs were posted per month

DA_totals = df_DA_US.groupby("job_posted_month_no").size()
```

```python
# calculate the % of skill over the total job postings in each month

df_DA_US_percent = df_DA_US_pivot.div(DA_totals/100, axis=0)
```

```python
# change the month number to a month name

df_DA_US_percent = df_DA_US_percent.reset_index()
df_DA_US_percent["job_posted_month"] = df_DA_US_percent["job_posted_month_no"].apply(lambda x: pd.to_datetime(x, format="%m").strftime("%b"))
df_DA_US_percent = df_DA_US_percent.set_index("job_posted_month")
df_DA_US_percent = df_DA_US_percent.drop(columns="job_posted_month_no")
df_DA_US_percent
```

### Plotting

```python
from matplotlib.ticker import PercentFormatter

df_plot = df_DA_US_percent.iloc[:, :5]
sns.lineplot(data=df_plot, dashes=False, legend='full', palette='tab10')
sns.set_theme(style='ticks')
sns.despine() # remove top and right spines

plt.title('Trending Top Skills for Data Analysts in the US')
plt.ylabel('Likelihood in Job Posting')
plt.xlabel('2023')
plt.gca().yaxis.set_major_formatter(PercentFormatter(decimals=0))
plt.legend(
    title="Job Skills",
    loc="upper center",
    bbox_to_anchor=(0.5, -0.15),
    ncol=5
)

plt.tight_layout()
plt.show()
```
