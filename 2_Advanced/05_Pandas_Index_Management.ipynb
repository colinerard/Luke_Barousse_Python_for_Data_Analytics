# 5 Pandas Index Management

```python
# import libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# load data
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset["train"].to_pandas()

# clean data - date column
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
```

```python
# let's investigate the index of our df

df.index
```

```python
# currently the index does not have a name

df.index.name
```

```python
# we can assign one 

df.index.name = "job_index"
```

```python
df.head()

# we now have a title above the index column
```

```python
# let's look at the data type -» integer

df.index.dtype
```

## 5.1 Three methods to cover

1. df.reset_index()
2. df.set_index()
3. df.sort_index()

### reset_index

```python
# reset_index()
# let's create a new df for jobs in the us

df_usa = df[df["job_country"] == "United States"]

df_usa
# we now have an index that is all over the place!
```

```python
# we can use the reset index with a method
# inplace = True means that we replace the current df rather than creating a new one

df_usa.reset_index(inplace=True)

df_usa
```

### set_index

Move an existing column to the index position.

```python
# let's say we want to move back to the job_index column as our index

df_usa.set_index("job_index", inplace=True)

df_usa

# the job_index column is now back at the index position
```

### sort_index

Hardly ever a case with a normal dataframe. But use cases for that when grouping or pivoting data

```python
median_pivot = df_usa.pivot_table(
    values="salary_year_avg",
    index="job_title_short",
    aggfunc=["median", "max", "min"]
    )

median_pivot
```

```python
# we can use the sort_values() method to sort the pivot table

median_pivot.sort_values(by=("median", "salary_year_avg"), ascending=False)
```

```python
# we can also use the sort_index to sort the pivot table

median_pivot.sort_index(inplace=True) # inplace=True to replace the previous pivot

median_pivot
```
