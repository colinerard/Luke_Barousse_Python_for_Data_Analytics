## 3.2 How to "copy data"

### What NOT TO DO

```python
# let's create an original copy of our dataset

dataset = load_dataset("lukebarousse/data_jobs")
df_original = dataset["train"].to_pandas()
```

```python
# to create a copy of the dataset, we could use this simple equation (but it as negative consequences!!)

df_altered = df_original
```

```python
# let's look at the first 5 rows to see what we have in the salary column

df_altered.loc[:5, "salary_year_avg"]
```

```python
# what's the median salary

median_salary = df_altered["salary_year_avg"].median()

median_salary
```

```python
# let's fill the NaN values with the median salary and see if it worked

df_altered["salary_year_avg"] = df_altered["salary_year_avg"].fillna(median_salary)

df_altered.loc[:5, "salary_year_avg"]
```

```python
# now let's compare with the values from the df_original

df_original.loc[:5, "salary_year_avg"]
```

```python
# THEY ARE THE SAME!!! Because we used the simple equation at the beginning (df_altered = df_original) using the assignment operator
# Both df IDs are now identical and changes made to either of them actually are changes to the same DF!!

print(id(df_original))
print(id(df_altered))
```

Create a copy of the DataFrame and drop rows where the salary_year_avg column has missing values. Display the first 5 rows of the cleaned DataFrame.

## 2.3.2

Create a new DataFrame that is a random sample of 10% of the original DataFrame. Display the first 5 rows of the sampled DataFrame.

```python
# calculate 10% of original dataframe
df_random_tenth_length = int(.1 * len(df_original))

# create the new df
df_copy_2 = df_original.sample(df_random_tenth_length)

# display the first 5 rows
df_copy_2.head()
```

```python
# easier method

# create the new DF
df_copy_2_2 = df_original.sample(frac=.1)

# first 5 rows
df_copy_2_2.head()
```
