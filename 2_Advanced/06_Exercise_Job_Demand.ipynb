Combine knowledge of pivot tables and graph plotting to illustrate in a line chart the number of monthly job postings in the usa across 3 job titles with the most amount of job postings overall.

```python
# import libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# load data
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset["train"].to_pandas()

# clean data - date column
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
```

```python
# create a US specific DF

df_usa = df[df["job_country"] == "United States"].copy()

df_usa
```

```python
# create a new column with the month data, using a month verbage e.g. January, February, etc.

df_usa["job_posted_month"] = df_usa["job_posted_date"].dt.strftime("%B")

df_usa
```

```python
# let's pivot our data in a format to plot it

df_usa_pivot = df_usa.pivot_table(
    index="job_posted_month",
    columns="job_title_short",
    aggfunc="size"
)
```

```python
# to have a proper pivot, we need to be able to sort by the month in chronological order (not alphabetical)
# we need to add a column with a month_no to help us sort

df_usa_pivot.reset_index(inplace=True) #-» reset the current index 
df_usa_pivot["month_number"] = pd.to_datetime(df_usa_pivot["job_posted_month"], format="%B").dt.month # add a month number column to have a number for each month named in the job posted month column
df_usa_pivot.sort_values("month_number", inplace=True) # sort the pivot based on the new month number column
df_usa_pivot.set_index("job_posted_month", inplace=True) # set job_posted_month as the index given the new order. We now have them sorted the way we want.
df_usa_pivot.drop(columns="month_number", inplace=True) # drop the month number column that we don't need to have in our pivot.
```

```python
df_usa_pivot

# everything is just as we want it!
```

```python
# we can now plot a line chart based on the pivot table above

df_usa_pivot.plot(kind="line")

# but we still have all the jobs included
```

```python
# we just want the jobs with the most amount of postings
# we can create a list variable that we then use to filter for

top_3 = df_usa["job_title_short"].value_counts().head(3)

top_3
```

```python
# we can use the index function + the tolist() method to make this info in a list that we use to filter our line chart

top_3 = top_3.index.to_list()
```

```python
# now the new plot

df_usa_pivot[top_3].plot(kind="line")
plt.title("Monthly Job Postings in the US for Top Data Jobs")
plt.xlabel("2023")
plt.ylabel("Job Count")
plt.legend() # doing  that removes the ugly legend title
plt.show()
```
