# 8 Pandas Concat Dataframes

Bring together different files (that ideally all have the same format) e.g. getting monthly reports that we want to aggregate in a master file.

If they don't have the same format, the new columns will be added to the right of the table and be empty for records from dataframes where these columns don't exist.

```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## Simple example first (with dummy data)

```python
# Sample dataset of job postings in January
job_postings_jan = pd.DataFrame({
    'job_id': [1, 2, 3, 4, 5],
    'job_title': ['Data Scientist', 'Data Analyst', 'Machine Learning Engineer', 'Data Scientist', 'Data Engineer'],
    'company': ['Company A', 'Company B', 'Company C', 'Company D', 'Company E'],
    'job_posted_date': pd.to_datetime(['2024-01-02', '2024-01-07', '2024-01-14', '2024-01-19', '2024-01-24'])
})

job_postings_jan
```

```python
# Sample dataset of job postings in February
job_postings_feb = pd.DataFrame({
    'job_id': [6, 7, 8, 9, 10],
    'job_title': ['Data Scientist', 'Data Analyst', 'Machine Learning Engineer', 'Data Scientist', 'Data Engineer'],
    'company': ['Company F', 'Company G', 'Company H', 'Company I', 'Company J'],
    'job_posted_date': pd.to_datetime(['2024-02-05', '2024-02-09', '2024-02-12', '2024-02-18', '2024-02-22'])
})

job_postings_feb
```

```python
# pd.concat() can concat dataframes

pd.concat([job_postings_jan, job_postings_feb])

# but the indexes are messed up
```

```python
# here's how to solve the index

pd.concat([job_postings_jan, job_postings_feb], ignore_index=True)
```

## Example with the actual job postings dataset

Let's create fake data to exercise concat. Let's use the original dataset and create a couple of new datasets per month to then concat back together.

```python
# create a month column

df["job_posted_month"] = df["job_posted_date"].dt.strftime("%b")

df
```

```python
# create 12 different dfs, one per month
# let's store it within a dictionary

# 1st step: create a list of all the different months

months = df["job_posted_month"].unique()

# 2nd step - beginning dictionary creation. The first "month" is the key, the second "month" is the value and then it's cycling through the months variable

dict_months = {month: df[df["job_posted_month"] == month] for month in months}

# we now have individual dataframes for each month. E.g. for jan:

dict_months["Jan"]
```

Let's now concat all the data for Q1 2023.

```python
df_Q1 = pd.concat([dict_months["Jan"], dict_months["Feb"], dict_months["Mar"]], ignore_index=True)

df_Q1
```

## 2.8.1

Concatenate two DataFrames to analyze job postings for Data Scientists and Data Analysts:

- Create one DataFrame with job postings for Data Scientists named data_scientist_df and another for Data Analysts named data_analyst_df.
- Concatenate them and display the combined DataFrame called combined_df.

```python
# Create data_scientist_df & data_analyst_df

data_scientist_df = df[df["job_title_short"] == "Data Scientist"].copy()

data_analyst_df = df[df["job_title_short"] == "Data Analyst"].copy()

# concatenate both dfs

combined_df = pd.concat([data_scientist_df, data_analyst_df]).reset_index()

combined_df
```

## 2.8.2

Concatenate DataFrames to analyze job postings by month:

- Create DataFrames for job postings for the last quarter (October, November, and December). These are respectively called oct_df, nov_df, and dec_df.
- Concatenate them and display the combined DataFrame named combined_df.

```python
# creating new df to avoid confusion
df2 = df.copy()

# add date column ()
df2["job_posted_month2"] = df2["job_posted_date"].dt.strftime("%b")

# create a list of unique months

months2 = df2["job_posted_month2"].unique()

# create unique df for each months with a loop and store them in a dictionary

dict_months2 = {month2: df2[df2["job_posted_month2"] == month2] for month2 in months2}

# concat
q4_2023_df2 = pd.concat([dict_months2["Oct"], dict_months2["Nov"], dict_months2["Dec"]], ignore_index=True)

q4_2023_df2
```

## 2.8.3

Concatenate DataFrames to analyze job postings for different job types:

- Create DataFrames for 'Full-time' roles (called full_time_df) and 'Part-time' roles (called part_time_df) job postings.
- Concatenate them and display the combined DataFrame named combined_df.

```python
# create dfs

full_time_df = df[df["job_schedule_type"] == "Full-time"].copy()

part_time_df = df[df["job_schedule_type"] == "Part-time"].copy()

# concat

combined_schedule_df = pd.concat([full_time_df, part_time_df], ignore_index=True)

combined_schedule_df
```
